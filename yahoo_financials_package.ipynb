```python
# Initial import
from yahoofinancials import YahooFinancials
import pandas as pd

# YahooFinancials docs and Github
# https://pypi.org/project/yahoofinancials/
# https://github.com/JECSand/yahoofinancials
```

```python
# Declare variable wth list of tickers to import various financial data
momentum_stocks = ['TSLA', 'GME', 'FCEL', 'SOLO', 'COTY', 'PLTR', 'PLUG', 'KOS', 'NIO', 'IDEX', 'XPEV', 'M']

# Define variable for data import
yahoo_financials_momentum = YahooFinancials(momentum_stocks)
```

```python
# Import Balance Sheet data for selected stock tickers

# Define variable to get data. The data returns a .json dictionary of dictonaries
yahoo_balance_sheet = yahoo_financials_momentum.get_financial_stmts('annual', 'balance')

# Create a variable for the .json data header
yahoo_bal_sheet_hist = yahoo_balance_sheet['balanceSheetHistory']
```

```python
# Define a function to do a nested loop to brng data into a dataframe. NOTE: The dataframe will be multi-indexed
balance_sheet_history = {}

for k in yahoo_bal_sheet_hist.keys():
    bs_stocks = yahoo_bal_sheet_hist[k]
    bs_periods_dict = {}
    for x in range(0, len(bs_stocks)):
        bs_periods_dict.update(bs_stocks[x])
        balance_sheet_history[k] = bs_periods_dict
```

```python
# Create Multi-Indexed DataFrame

# Define a variable and function for the nested loop to bring the data into a DataFrame
balance_sheet = {(outerKey, innerKey): values for outerKey, innerDict in balance_sheet_history.items() for innerKey, values in innerDict.items()}

# Create DataFrame
balance_sheet_df = pd.DataFrame(balance_sheet)

# Display DataFrame
balance_sheet_df
```

```python
# Create Multi-Indexed DataFrame

# Define a variable and function for the nested loop to bring the data into a DataFrame
income_statement = {(outerKey, innerKey): values for outerKey, innerDict in income_statement_history.items() for innerKey, values in innerDict.items()}

# Create DataFrame
income_statement_df = pd.DataFrame(income_statement)

# Display DataFrame
income_statement_df
```

```python
# Import Quarterly Cash Flows data for selected stock tickers

# Define variable to get data. The data returns a .json dictionary of dictonaries
yahoo_daily_price_data = yahoo_financials_momentum.get_historical_price_data("2020-12-01", "2020-12-31", "daily")
```

```python
single_ticker_df = pd.DataFrame(yahoo_daily_price_data['TSLA']['prices'])
single_ticker_df['currency'] = yahoo_daily_price_data['TSLA']['currency']
single_ticker_df['instrumentType'] = yahoo_daily_price_data['TSLA']['instrumentType']
single_ticker_df 
```

```python
# Define a function to do a nested loop to brng data into a dataframe. NOTE: The dataframe will be multi-indexed
ticker_list = list(yahoo_daily_price_data.keys())
df_list = []
success_list = []
for each_ticker in ticker_list: 
    print(f'Getting {each_ticker}')
    try:
        single_ticker_df = pd.DataFrame(yahoo_daily_price_data[each_ticker]['prices'])
        single_ticker_df['currency'] = yahoo_daily_price_data[each_ticker]['currency']
        single_ticker_df['instrumentType'] = yahoo_daily_price_data[each_ticker]['instrumentType']
        success_list.append(each_ticker)
        df_list.append(single_ticker_df)
    except: 
        print(f'Issue with {each_ticker}')
single_ticker_df
```

```python
# Create Multi-Indexed DataFrame

multi_index=[]
data_dict={}
for idx, ticker in enumerate(success_list):
    for each_column in df_list[idx].columns: 
        multi_index.append((ticker, each_column))
        data_dict[(ticker, each_column)] = df_list[idx][each_column]

hist_ticker_price_df = pd.DataFrame(data_dict)
hist_ticker_price_df.head()
```
